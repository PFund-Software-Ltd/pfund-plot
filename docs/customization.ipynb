# Customization

If you want to customize the plot, you can set `raw_figure` to True to get the raw figure object.
In this example, a bokeh figure is returned.

```python
import pandas as pd
import pfund_plot as plt
from bokeh.sampledata import stocks
from bokeh.plotting import figure
from bokeh.io import show

df = pd.DataFrame(stocks.AAPL)

fig: figure = plt.candlestick(df, display_mode='notebook', raw_figure=True)
fig
```

## Default figure:

```python
show(fig)
```

## Customized figure:

```python
# change the title
fig.title.text = "Customized Candlestick Plot"
fig.title.text_font_size = '20pt'
fig.title.text_color = 'blue'
show(fig)
```
